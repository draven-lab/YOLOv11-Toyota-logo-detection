```python
from ultralytics import YOLO

# 訓練自己的模型
model = YOLO('yolo11n.pt')  # 可用 yolov8n/yolov11n 做起點

model.train(
    data = 'data.yaml',  # 修改成你的路徑
    epochs = 100,     # 建議 50~100，看 loss 曲線收斂即可
    imgsz = 640,
    batch = 8,        # 依顯卡調整
    workers = 2,
    project = 'toyota_logo_project',
    name = 'exp12',
    device='cuda'
)
```

```python
# 驗證 (optional)
metrics = model.val()
print(metrics)
```

```python
# 影片檔案路徑
video_path = r'C:\Users\user\OneDrive\桌面\yolov11\toyota.mp4'

# 載入訓練好的模型
model = YOLO(r"C:\Users\user\OneDrive\桌面\yolov11\toyota_logo_project\exp12\weights\best.pt")  # 修改成你的最佳權重路徑

# 推論並儲存影片
results = model.track(
    source=video_path, 
    show=False,      # 如果在 Colab 建議設 False
    save=True, 
    conf=0.25,
    iou=0.5,
    classes=0        # 只偵測 toyota 類別（假設只有一類）
)
```
